```python
import re
import pandas as pd
```

```python
f = open('WhatsApp Chat with MITHHU GANG.txt', 'r', encoding='utf-8')
```

```python
data = f.read()
```

```python
print(data) # -->data is stored in the form of the string
```

```python
pattern = '\d{1,2}/\d{1,2}/\d{2,4},\s\d{1,2}:\d{2}\s(?:AM|PM)\s-\s'
# 3/21/22, 11:12â€¯PM - Prathmesh Bhoir: Hii --> we the string in the regular expression
# for that we use the website regex101.com
```

```python
messages = re.split(pattern, data)[1:]
messages
```

```python
dates = [date.replace("\u202f", " ").replace(" -", "").strip() for date in re.findall(pattern, data)]
dates
```

```python
df = pd.DataFrame({'user_message':messages, 'message_date': dates})
# convert message_date type
df['message_date'] = pd.to_datetime(df['message_date'].str.strip(), format='%m/%d/%y, %I:%M %p')

df.rename(columns={'message_date': 'date'}, inplace=True)

df.head()
```

```python
df.shape
```

```python
# separate users and messages
users = []
messages = []

for message in df['user_message']:
    entry = re.split(r'^([\w\s]+?):\s', message)  # FIXED regex
    if len(entry) > 2: 
        users.append(entry[1])  # Full name
        messages.append(entry[2])  # Message content
    else:
        users.append('group_notification')
        messages.append(entry[0])

df['user'] = users
df['message'] = messages
df.drop(columns=['user_message'], inplace=True)

df.head()
```

```python
period = []
for hour in df[['day_name', 'hour']]['hour']:
    if hour == 23:
        period.append(str(hour) + "-" + str('00'))
    elif hour == 0:
        period.append(str('00') + "-" + str(hour+1))
    else:
        period.append(str(hour) + "-" + str(hour+1))
```

```python
df['period'] = period
```
